Read student data into a list from a csv file:

    loop through the list and print each student with name, img_url and avg_grade.
    sort the list by avg_grade

```python
import csv

file_name = "studens.csv"
with open(file_name) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    students = []
    for row in reader:
        name, course_name, teacher, ects, room, grade, url, gender = row
        course = Course( course_name, room, teacher, ects, int(grade))
        data_sheed = Data_sheed([course])
        students.append(Student(data_sheed, name, gender, url))
    sort_stud = sorted(students, key=Student.get_avg_grade)
    for student in sort_stud:
        
        print(student.name, student.image_url, student.get_avg_grade())
        
```

If list is shorter than 3 raise your own custom exception (NotEnoughStudentsException)

```python
def furthest_students(students):
    """findes the 3 students that are furthest with their studies. sorted on etcs points"""
    sort_stud = sorted(students, key=Student.show_prog)
    if(len(sort_stud)<3):
        raise NotEnoughStudentsException("you supplied: ",len(sort_stud)," but 3 is needed" )
    return sort_stud[-3:]
   

#students is a list of random students with a bunsh of courses
students = generate_rand_students_for_exceptions(4)
furthest_students(students)
print(students)
```
