```python
import numpy as np
import pandas as pd

data = pd.read_csv("titanic/train.csv")
data.head()
```

```python
data = data.drop(columns=["PassengerId", "Name","Ticket","Cabin"])
data.head()
```

```python
from sklearn import preprocessing

label_enc = preprocessing.LabelEncoder()
data["Sex"] = label_enc.fit_transform(data["Sex"].astype(str))
print(data.head())
```

```python
data = pd.get_dummies(data, columns=['Embarked'])
print(data)
```

```python
print("rows before drop: ", len(data))

null_matrix = data.isnull()

only_null_filter = null_matrix.any(axis=1)

missing = data[only_null_filter]

data_not_null = data.dropna()

print(len(data_not_null))
```

```python
from sklearn.cluster import estimate_bandwidth
estimate_bandwidth(data_not_null)
```

```python
from sklearn.cluster import MeanShift
analyzer = MeanShift(bandwidth=30)
analyzer.fit(data_not_null)
```

```python
labels = analyzer.labels_
print(labels)
```

```python
data_not_null["cluster_group"] = np.nan

data_len = len(data_not_null)

for i in range(data_len):
    data_not_null.iloc[i,data_not_null.columns.get_loc("cluster_group")] = labels[i]

```

```python
data_not_null.describe()
```

```python
cluster_titanic = data_not_null.groupby(["cluster_group"]).mean()


#cluster_titanic['Counts'] = pd.Series(data_not_null.groupby(['cluster_group']).size())
cluster_titanic
```
